```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as myplot
```

```python
myData=pd.read_csv('Position_Salaries.csv')
myData
```

```python
x= myData.iloc[:,1:2].values
x
```

```python
y = myData.iloc[:,2].values
y
```

```python
#fitting the regression model to datasets
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
```

```python
y_pred = regressor.predict(6.5)
y_pred
```

```python
myplot.scatter(x,y,color='r')
myplot.plot(x,regressor.predict(x),color='b')
myplot.title('truth of bluf')
myplot.xlabel('position level')
myplot.ylabel('salary')
myplot.show()
```

```python
x_grid = np.arange (min(x),max(x),0.01)
x_grid = x_grid.reshape(len(x_grid),1)
x_grid

```

```python
myplot.scatter(x,y,color='r')
myplot.plot(x_grid,regressor.predict(x_grid),color='b')
myplot.show()
```
