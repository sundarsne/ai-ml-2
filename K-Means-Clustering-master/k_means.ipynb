```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%config InlineBackend.figure_format  ='retina'
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
np.random.seed(5)
iris = datasets.load_iris()
x=iris.data
x
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters =i,init = 'k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) 
plt.plot(range(1,11),wcss)
plt.title('the elbow method')
plt.xlabel('no.of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
estimators =[('k_means_iris_8',KMeans(n_clusters= 8, init='k-means++')),
             ('k_means_iris_3',KMeans(n_clusters= 3, init='k-means++',)),
              ('k_means_iris_bad_init',KMeans(n_clusters= 3,n_init= 1,
                                             init = 'random'))] 
fignum =1 
titles = ['8 clusters','3 clusters','3 cluster,bad intialisation']
for name ,estimator in estimators:
    fig = plt.figure(fignum,figsize=[7,5])
    ax = Axes3D(fig,rect=[0, 0, .95, 1],elev=48,azim=134)
    estimator.fit(x)
    labels=estimator.labels_
    
    ax.scatter(x[:,3],x[:,0],x[:,1],
                  c=labels.astype(np.float),edgecolor='k')
    
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('petal width')
    ax.set_ylabel('sepal length')
    ax.set_zlabel('petal length')
    ax.set_title(titles[fignum-1])
    ax.dist=12
    fignum=fignum+1
```

```python
y =iris.target
fig = plt.figure(fignum,figsize=(7,5))
ax = Axes3D(fig,rect=[0,0,.95,1],elev=48,azim=134)
for name,label in[('setosa',0),
                  ('versicolor',1),
                  ('verginica',2)]:
    ax.text3D(x[y == label,3].mean(),
              x[y == label,0].mean(),
              x[y == label,1].mean()+2,name,
            horizontalalignment='center',
            bbox = dict(alpha=.2,edgecolor='w',facecolor='w'))
y=np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(x[:,3],x[:,0],x[:,2],c=y,edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('petal width')
ax.set_ylabel('sepal length')
ax.set_zlabel('sepal length')
ax.set_title('ground truth')
ax.dist=12
fig.show()
```
